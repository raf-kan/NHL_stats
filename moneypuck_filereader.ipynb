```python
def x_per_y(data, player, x, y):

    type_dict = {
                # Y Values
                'game':'games_played',
                'shift' : 'shifts',
                'ice time': 'icetime',
                'bench': 'timeOnBench',
                'penalty minutes': 'penalityMinutes',
                # X Values
                'goals': 'I_F_goals',
                'points': 'I_F_points',
                'shot attempts': 'I_F_shotAttempts',
                'penalties': 'penalties',
                'faceoffs won': 'faceoffsWon',
                'faceoffs lost': 'faceoffsLost',
                'hits': 'I_F_hits',
                'take aways': 'I_F_takeaways',
                'give aways': 'I_F_giveaways',
                'penalties': 'penaltiesDrawn'
                }
    
    if y in type_dict.keys():
        y = type_dict[y]
    else:
        raise ValueError
    
    if x in type_dict.keys():
        x = type_dict[x]
    else:
        raise ValueError

    try:
        player_data = data.loc[data['name']==player]
    except:
        raise ValueError ('Player not found')
    x_val = int(player_data[x])
    y_val = int(player_data[y])

    print(f'{x}: {x_val}')
    print(f'{y}: {y_val}')
    return x_val/y_val
    # return type(player_data)
```

```python
print(x_per_y(skaters_all, player = 'Jordan Kyrou', x = 'points', y = 'game'))
```
